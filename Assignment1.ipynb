```python
x = np.array([2,4,6,8,10])
y = np.array([3,7,5,10,12])
n = len(x)
sx = np.sum(x)
sy = np.sum(y)
sxy = np.sum(x * y)
sx2 = np.sum(x ** 2)
slope = (n * sxy - sx * sy) / (n * sx2 - sx ** 2)
intercept = (sy - slope * sx) / n
print(slope)
print(intercept)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Input features (Square Feet) and target values (Price in arbitrary units)
X = np.array([2, 4, 6, 8, 10]).reshape(-1, 1)
Y = np.array([3, 7, 5, 10, 12])

# Initialize and fit the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict values using the trained model
Y_pred = model.predict(X)

# Visualizing the data points and the regression line
plt.scatter(X, Y, color='blue', label='Actual Prices')
plt.plot(X, Y_pred, color='red', label='Predicted Prices')
plt.title('Linear Regression: House Price Prediction')
plt.xlabel('Square Feet (x1000)')
plt.ylabel('Price (in Lakhs)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import math
from collections import Counter

data = {
        "Year" : [2018,2019,2020,2021,2022,2023,2024],
        "Market_Competition" : ['Medium','High','Low','Medium','High','Low','High'],
        "Type" : ['A','B','A','B','A','A','B'],
        "Profit_Loss" : ['Profit','Profit','Loss','Profit','Profit','Loss','Loss']
}

df = pd.DataFrame(data)

def entropy(values):
    total = len(values)
    counts = Counter(values)
    entropy_val = 0

    for count in counts.values():
        p = count / total
        entropy_val -= p * math.log2(p)

    return entropy_val

profit_loss_entropy = entropy(df["Profit_Loss"])
print(profit_loss_entropy)

def information_gain(df,feature,target):
    total_entropy = entropy(df[target])
    unique_values = df[feature].unique()
    weighted_entropy = 0
    for value in unique_values:
        subset = df[df[feature]==value][target]
        subset_entropy = entropy(subset)
        weighted_entropy += (len(subset)/len(df))*subset_entropy
    return total_entropy - weighted_entropy

market_competition = information_gain(df,"Market_Competition", "Profit_Loss")
year = information_gain(df,"Year","Profit_Loss")
print(market_competition)
print(year)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.preprocessing import LabelEncoder

# Given data (ensuring all lists have the same length)
data= {
    "Year" : [2018, 2019, 2020, 2021, 2022],
    "Market_Competition" : ['Medium', 'High', 'Low', 'Medium', 'High'],
    "Type" : ['A', 'B', 'A', 'B', 'A'],
    "Profit_Loss" : ['Profit', 'Profit', 'Loss', 'Profit', 'Profit']
}
# Creating DataFrame
df = pd.DataFrame(data)

# Encoding categorical variables
le_market = LabelEncoder()
le_type = LabelEncoder()
le_profit = LabelEncoder()

df['Market_Competition'] = le_market.fit_transform(df['Market_Competition'])
df['Type'] = le_type.fit_transform(df['Type'])
df['Profit_Loss'] = le_profit.fit_transform(df['Profit_Loss'])  # Target variable

# Splitting features and target
X = df[['Year', 'Market_Competition', 'Type']]
y = df['Profit_Loss']

# Creating Decision Tree model
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
clf.fit(X, y)

# Plotting the Decision Tree
#plt.figure(figsize=(10, 6))
tree.plot_tree(clf, feature_names=X.columns, class_names=le_profit.classes_)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Create the dataset
data = {
    "Student": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"],
    "Math": [85, 70, 60, 95, 55, 80, 45, 90, 78, 65],
    "Science": [90, 75, 65, 92, 50, 85, 40, 88, 82, 60],
    "English": [80, 85, 70, 96, 60, 88, 50, 86, 80, 72]
}

df = pd.DataFrame(data)

# Step 2: Select features for clustering
X = df[["Math", "Science", "English"]]

# Step 3: Apply KMeans clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X)

# Step 4: Show the students and their cluster assignments
print(df[["Student", "Cluster"]])

# Step 5: Visualize the clusters (using Math vs Science for 2D plot)
plt.scatter(df["Math"], df["Science"], c=df["Cluster"])
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.title("K-Means Clustering of Students Based on Marks")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Step 1: Define the data
data = {
    "Student": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"],
    "Math": [85, 78, 90, 65, 55, 89, 76, 92, 60, 58],
    "Science": [80, 75, 85, 60, 50, 87, 74, 95, 55, 52]
}

# Step 2: Create a DataFrame
df = pd.DataFrame(data)

# Step 3: Select features for clustering
X = df[["Math", "Science"]]

# Step 4: Apply Agglomerative Clustering with 3 clusters
agglo = AgglomerativeClustering(n_clusters=3)
df["Cluster"] = agglo.fit_predict(X)

# Step 5: Print cluster assignments
print(df[["Student", "Cluster"]])

# Step 6: Visualize the clusters
plt.scatter(df["Math"], df["Science"], c=df["Cluster"])
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.title("Agglomerative Clustering of Students")
plt.show()
```
