```python
matrix1 = [[1,2,3],[4,5,6]]

def transpose_matrix(matrix):
    transpose = []
    for i in range(len(matrix[0])): # no. of columns
        r = []
        for j in range(len(matrix)): # no. of rows
            r.append(matrix[j][i])
        transpose.append(r)
    return transpose

transpose_mat = transpose_matrix(matrix1)
print(transpose_mat)
```

```python
import numpy as np
import math

states = ['Sunny','Rainy', 'Cloudy']
matrix = np.array([[0.6,0.2,0.2],[0.3,0.5,0.2],[0.5,0.4,0.1]])
def func1(csi):
    prob = matrix[csi]
    nsi = np.random.choice([0,1,2],p=prob)
    return nsi
def func2(steps):
    csi = np.random.choice([0,1,2])
    sequence = [states[csi]]
    for i in range(steps-1):
        csi = func1(csi)
        sequence.append(states[csi])
    return sequence
f2=func2(12)
print(f2)
```

Q10

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
import numpy as np

y_true = np.array([0,1,2,2,0,1,1,2,0,0])
y_pred = np.array([0,1,2,1,0,2,1,2,0,1])

report  = classification_report(y_true, y_pred)
print(report)

classes = np.unique(y_true)
precision_cls = []
recall_cls = []
f1_cls = []

for cls in classes:
    TP = np.sum((y_pred == cls)&(y_true == cls))
    FP = np.sum((y_pred == cls)&(y_true != cls))
    FN = np.sum((y_pred != cls)&(y_true == cls))

    precision = TP/(TP+FP) if (TP+FP)!=0 else 0
    recall = TP/(TP+FN) if (TP+FN)!=0 else 0
    F1_score = (2*precision*recall)/(precision+recall) if (precision+recall)!=0 else 0

    precision_cls.append(precision)
    recall_cls.append(recall)
    f1_cls.append(F1_score)

plt.figure(figsize=(8,5))
plt.bar(classes,precision_cls,color='blue')
plt.bar(classes,recall_cls,color='red')
plt.bar(classes,f1_cls,color='yellow')
plt.xlabel('Classes')
plt.ylabel('Scores')
plt.title("Precision,Recall,F1 Score")
plt.show()
```

```python
import numpy as np

y_true = np.array([4.5,1.5,11.4,13.3,11.0,10.9])
y_pred = np.array([3.4,2.1,12.1,10.2,12.0,14.2])

sum_squared_error = 0
for i in range(len(y_true)):
    diff = y_true[i] - y_pred[i]
    sum_squared_error += diff ** 2

mse = sum_squared_error/len(y_true)
print(mse)
```

```python
import numpy as np
import math
y_true = np.array([1,0,1,1])
y_pred = np.array([0.9,0.2,0.8,0.7])

bce_loss = 0
for i in range(len(y_true)):
    bce_loss += -(y_true[i]*math.log(y_pred[i])+(1-y_true[i])*math.log(1-y_pred[i]))

bce_loss /= len(y_true)
print(bce_loss)
```

```python
import numpy as np

#given true values(one hot encoded) & pred problem
y_true=np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
y_pred=np.array([[0.7, 0.2, 0.1], [0.1, 0.8, 0.1], [0.2, 0.1, 0.7]])

#function to calculate the multi-class entropy loss
def multiclass_cross_entropy(y_true,y_pred):
#clip predictions to avoid log(0) errors
    epsilon=1e-15
    y_pred=np.clip(y_pred,epsilon,1-epsilon)
    loss=-np.mean(np.sum(y_true*np.log(y_pred),axis=1))
    return loss
loss=multiclass_cross_entropy(y_true, y_pred)
print(loss)
```
